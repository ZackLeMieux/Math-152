#Instructions:


Begin with a positive integer n. If it is even, divide it by two. If it is odd, replace it by
3n+1. Then repeat. This process leads to fascinating ups and downs, that mathematicians
struggle to understand. For example, if we start with 5, then the next number will be
3 ·5 + 1 = 16. The next number will be 8, then 4, then 2, then 1. The next number will
be 3 ·1 + 1 = 4. Then 2. Then 1. In this way, the starting number 5 yields the sequence:

5,16,8,4,2,1,4,2,1,4,2,1,...

It is conjectured (“The Collatz conjecture”) that if you start with any positive integer
n, the resulting sequence will eventually reach the “cycle” 4,2,1,4,2,1. But nobody can
prove it!
Please explore this conjecture in your notebook.

```python
def collatz(n):               #First Function to code the Collatz Conjecture.
  while n!=1:                 #Returns the values of the sequence for a given input.
    if n%2==0:
     n=n//2
     print(n)
    else:
     n=3*n+1
     print(n)
  else:
    print(n)
  return n
```

```python
collatz(7)
```

```python
def collatz2(n):           # This function will loop from 2 to the given integer to check all integers within the range
  for x in range(2,n+1):   # This for loop allows us to cycle up to the inputed integer.
    x_1=x                  # x_1 is the dummy integer that is being tested to see if it follows the collatz conjecture
    counter=0              # The counter will be used to check for repetition of the cycle.
    while x_1 !=1:         # The While loop allows for repeated looping until the condition is met.
      if x_1%2==0:         # The If condition is checking for evenness of the dummy variable.
        x_1=x_1//2         # If the dummy variable is even this line divides it by half.
        counter+=1         # This counter counts all opperations on the dummy variable
      elif x_1%2==1:       # The Elif condition checks for oddness of the dummy variable.
        x_1=3*x_1+1        # If the dummy variable is odd we multiply by 3 and add 1
        counter+=1         # This counter counts all opperations on the dummy variable.
      elif counter>100000: # The last Elif is crucial since it will be the way that we find any values that do not follow the conjecture.
        return(False)      # We get False if the counter reaches an arbitrarily large number, in this case 100,000
        print(x, 'breaks the conjecture!')     
      else:
        return(False)
        print(x, 'breaks the conjecture!')
    else:
      print('The number of steps for;', x ,'is', counter)
```

```python
collatz2(100)
```

```python
def collatz3(n):           
  for x in range(2,n+1):   
    x_1=x                  
    counter=0              
    while x_1 !=1:         
      if x_1%2==0:         
        while x_1%2==0:       #This is the first of two While statements added from collatz2 that will check to see if the next value is also even
          x_1=x_1//2         
          counter+=1      
      elif x_1%2==1:       
        while x_1%2==1:
          x_1=3*x_1+1        
          counter+=1 
      elif counter>100000: 
        return(False) 
        print(x, 'breaks the conjecture!')     
      else:
        return(False)
        print(x, 'breaks the conjecture!')
    else:
      print('The number of steps for;', x ,'is', counter)
```

```python
collatz3(100)
```

```python
def collatz4(n):           
  for x in range(2,n+1):   
    x_1=x                  
    counter=0              
    while x_1 !=1:         
      if x_1%2==0:         
        while x_1%2==0:       
          x_1=x_1//2         
          counter+=1      
      elif x_1%2==1:       
        while x_1%2==1:
          x_1=3*x_1+1        
          counter+=1                                   #Taking out the elif counter is greater than 100,000 condition does not seem to have an effect on the run time
    else:
      print('The number of steps for;', x ,'is', counter)
```

Something interesting happens when allowing the classical model to have as inputs negative numbers; there are two situations that have two unique cycles.

1. An input of (-6) would have a cylce of; (-6) -> -3,-8,-4,-2,-1,-2,-1,-2. Or in orther words we now cycle between -1 and -2 as opposed to the 4,2,1,4,2,1 cycle when only allowing positive integers.

2. The other situation, with an input of; (-5) -> -14,-7,-20,-10,-5,-14,-7,-20,-10,... which is a cycle that never reaches the integers of 2 or 1 and will cycle through these values indefinately.

3. In fact, it seems that this -14,-7,-20,-10,-5,-14 cycle is the only other cycle that a negative integer can get too, unless the integer catches a multiple of 8. 

Naturally it seems that the next function to write would be one that could output a cycle, giving us both the values of that cycle as well as the length.

```python
def neg_collatz(n):                             #First attempt at the code but was unable to stop the looping.
  counter=0
  x=n
  while n!=0 or abs(n)!=1 or abs(n)!=5:
    if n%2==0:
     n=n//2
     print(n)
     counter+=1
     print(counter)
     if x==n:
       return counter
       n==0
    else:
     n=3*n+1
     print(n)
     counter+=1
     print(counter)
     if x==n:
       return counter
       n==0
  return n
```
